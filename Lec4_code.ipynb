### Clean Protein Structure

```python
from Bio.PDB import PDBList, PDBParser, Select, PDBIO

# PDB ID 지정
pdb_id = "8YTY"  # 예시 PDB ID

# PDB 파일 다운로드
pdbl = PDBList()
pdb_file = pdbl.retrieve_pdb_file(pdb_id, pdir=".", file_format='pdb')

# PDB 구조 파싱
parser = PDBParser(QUIET=True)
structure = parser.get_structure(pdb_id, pdb_file)

# 단백질만 선택하는 클래스 정의
class ProteinOnlySelect(Select):
    def accept_residue(self, residue):
        # 잔기가 표준 아미노산이면 True
        return residue.get_id()[0] == ' '

# 전처리 후 저장
io = PDBIO()
io.set_structure(structure)
cleaned_pdb_filename = f"{pdb_id}_cleaned.pdb"
io.save(cleaned_pdb_filename, ProteinOnlySelect())

print(f"Cleaned PDB file saved as: {cleaned_pdb_filename}")
```
